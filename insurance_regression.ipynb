# Simple Linear Regression

Demonstration of the scikit-learn linear regression tools.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# Import the data
data = pd.read_csv("insurance.csv")
data.head()
```

```python
# Select data for plotting
age = data.iloc[:,0].values.reshape(-1,1)
charges = data.iloc[:,-1].values

# Plot data
plt.scatter(age,charges,color = 'b')
plt.xlabel('Age')
plt.ylabel('Insurance charges')
```

```python
# Apply the linear regression model and print coefficent
charges_model = LinearRegression()
charges_model.fit(age,charges)


print('Coefficient: \n', charges_model.coef_)
```

Use the linear regression model to predict insurance charges for a person who is 75 years old. 

```python
# Predict an unknown value
unk_age = [[75]]

age_pred = np.append(age, unk_age).reshape(-1,1)
charges_pred = charges_model.predict(age_pred)

plt.scatter(age,charges,color = 'b')
plt.plot(age_pred,charges_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Insurance charges')
plt.show()

print(f"Person with age: {unk_age} has expected charges of {charges_model.predict(unk_age)}")
```
